```python
import seaborn as sns
import matplotlib.font_manager as fm
import matplotlib.pyplot as plt
import matplotlib as mpl
# 2. นำฟอนต์เข้าสู่ระบบและตั้งค่าให้เป็นฟอนต์มาตรฐานของกราฟ
fm.fontManager.addfont('thsarabunnew-webfont.ttf')
mpl.rc('font', family='TH Sarabun New', size=14)

# --- ลำดับที่ 2: กราฟข้อ 1 (ใครได้รับค่าตอบแทนแค่ไหน?) ---
plt.figure(figsize=(12, 6))
# วาด Boxplot โดยให้แกน x คือรายได้, แกน y คืออาชีพ
# ใช้ Boxplot + ให้แสดงค่าเฉลี่ย (จุดสีแดง) + แยกสีตามกลุ่มอาชีพหลัก
sns.boxplot(
    data=df_total,
    x='value',
    y='soc_eco_class2',
    hue='soc_eco_class1',  # แยกสีตามกลุ่มหลัก (เจ้าของธุรกิจ, ลูกจ้าง, เกษตรกร)
    dodge=False,           # ไม่ต้องแยกกล่องย่อยถ้าแกน y เป็นอันเดียวกันแล้ว
    showmeans=True,        # *** สำคัญมาก: สั่งให้แสดง "ค่าเฉลี่ย" ด้วย ***
    meanprops={"marker":"o", "markerfacecolor":"red", "markeredgecolor":"black", "markersize":"8"} # ตั้งค่าจุดเฉลี่ยเป็นวงกลมสีแดง
)
# 5. ตกแต่งกราฟให้สื่อสารชัดเจน
plt.title('ข้อ 1: การกระจายตัว ช่องว่าง และรายได้เฉลี่ย จำแนกตามกลุ่มอาชีพย่อยและกลุ่มหลัก', fontsize=20, fontweight='bold')
plt.xlabel('รายได้ทั้งสิ้นต่อเดือน (บาท)', fontsize=16)
plt.ylabel('กลุ่มอาชีพย่อย', fontsize=16)

# ย้ายกล่องคำอธิบายสี (Legend) ไปไว้ที่ว่างๆ จะได้ไม่บังข้อมูล
plt.legend(title='ประเภทผู้ประกอบการ (กลุ่มหลัก)', bbox_to_anchor=(1.05, 1), loc='upper left')

plt.show()
```

```python
# 2. คัดกรองเอาเฉพาะยอดรวม "รายได้ทั้งสิ้นต่อเดือน"
df_total = df_clean[df_clean['source_income1'] == 'รายได้ทั้งสิ้นต่อเดือน'].copy()

# 3. จัดกลุ่มตามอาชีพย่อย แล้วสั่งให้คำนวณ "ค่าเฉลี่ย (mean)" ของรายได้
summary = df_total.groupby('soc_eco_class2')['value'].mean()

# 4. เรียงลำดับจากรายได้มากไปน้อย เพื่อให้ดูง่าย
summary_sorted = summary.sort_values(ascending=False)

# 5. แสดงผลลัพธ์
print("--- สรุปรายได้เฉลี่ยเรียงจากมากไปน้อย ---")
print(summary_sorted)
```

heatmap ดูทุกจังหวัด

```python
# 4. คำนวณความเหลื่อมล้ำเพื่อเอามาจัดเรียงลำดับ (จากมากไปน้อย)
std_series = df_total.groupby('province')['value'].std()

# 5. ทำ Pivot Table รวมทุกจังหวัด
pivot_df = df_total.pivot_table(index='province', columns='soc_eco_class1', values='value', aggfunc='mean')

# แอบเอาค่า SD มาแปะในตารางชั่วคราว เพื่อสั่งเรียงลำดับ แล้วค่อยลบออก
pivot_df['std'] = std_series
pivot_df = pivot_df.sort_values(by='std', ascending=False)
pivot_df = pivot_df.drop(columns=['std'])

# 6. วาด Heat Map (ปรับ figsize ให้สูงถึง 25 นิ้ว!)
plt.figure(figsize=(12, 25))

# วาดกราฟและย่อขนาดแถบสี (cbar_kws={'shrink': 0.5}) ให้ดูสมส่วนกับกราฟที่ยาวขึ้น
sns.heatmap(pivot_df, cmap='OrRd', annot=True, fmt=',.0f', linewidths=.5,
            cbar_kws={'label': 'รายได้เฉลี่ย (บาท)', 'shrink': 0.5})

plt.title('Heat Map แสดงระดับรายได้และช่องว่างของทั้ง 77 จังหวัด\n(เรียงลำดับจากจังหวัดที่เหลื่อมล้ำมากที่สุดไปน้อยที่สุด)',
          fontsize=20, fontweight='bold', pad=20)
plt.xlabel('ประเภทผู้ประกอบการ (กลุ่มหลัก)', fontsize=16)
plt.ylabel('จังหวัด', fontsize=16)
plt.xticks(rotation=15)

plt.tight_layout()
plt.show()
```

จังหวัดใดเป็นพื้นที่ที่มีความเหลื่อมล้ำสูงที่สุด = จันทบุรี

ทำ quandant analysisคือการ เอา 2 ตัวแปรเชิงปริมาณมา plot ตัดกันด้วยเส้น thresholdเพื่อแบ่งข้อมูลเป็น 4 กลุ่ม<br>ใช้มาตรวัด cv เพื่อใช้วัดค่าความแปรปรวนสัมพัทธ์ของรายได้ในแต่ละพื้นที่

ความจริงใช้มาตรวัด sd ก็ได้แต่มันมี bias จังหวัดที่มีรายได้สูงจึงใช้ มาตรวัด CV แทนเนื่องจาก ค่า CV ได้มาจาก CV = SD / mean จะทำให้เปรียบเทียบค่าเหลื่อมลำ้สัมพัทธ์ได้ยุติธรรมกับจังหวัดต่างๆได้ดีกว่า

```python
# สร้าง prov_stats: รายได้เฉลี่ย, SD, และ CV ของแต่ละจังหวัด
prov_stats = df_total.groupby('province')['value'].agg(
    mean='mean',
    std='std'
).reset_index()

# CV = SD / mean (ความแปรปรวนสัมพัทธ์)
prov_stats['cv'] = prov_stats['std'] / prov_stats['mean']

# ค่าเฉลี่ยระดับประเทศ สำหรับใช้ลาก threshold line
national_mean = prov_stats['mean'].mean()
national_cv   = prov_stats['cv'].mean()

prov_stats.head()
```

```python
# 3. สร้างกราฟ Scatter Plot (เปลี่ยนแกน Y เป็น CV)
# ขยายขนาดกราฟเป็น 16x12 เพื่อให้มีพื้นที่สำหรับใส่ชื่อทั้ง 77 จังหวัด
plt.figure(figsize=(16, 12))
sns.scatterplot(data=prov_stats, x='mean', y='cv', s=100, color='blue', alpha=0.6)

plt.axvline(national_mean, color='red', linestyle='--', label=f'รายได้เฉลี่ยประเทศ')
plt.axhline(national_cv, color='green', linestyle='--', label=f'ความเหลื่อมล้ำเฉลี่ยประเทศ (CV)')

# --- จุดที่แก้ไข: แปะชื่อ "ทุกจังหวัด" ---
for i, row in prov_stats.iterrows():
    # เราเอาคำสั่ง if ที่คอยกรองชื่อออกไป เพื่อให้มันพิมพ์ทุกแถว
    # และลด fontsize ลงเหลือ 10, ใส่ alpha=0.8 (โปร่งแสงนิดๆ) จะได้ไม่ทับจุด
    plt.text(row['mean']+200, row['cv']+0.005, row['province'], fontsize=10, alpha=0.8)

plt.title('ข้อ 2: การจัดกลุ่มจังหวัดเพื่อกำหนดนโยบาย (ใช้มาตรวัด CV - แสดงครบ 77 จังหวัด)', fontsize=20, fontweight='bold')
plt.xlabel('รายได้เฉลี่ยโดยรวมของจังหวัด (บาท)', fontsize=16)
plt.ylabel('ระดับความเหลื่อมล้ำ (CV = ความผันผวนเทียบกับฐานรายได้)', fontsize=16)
# ใส่ข้อความบอกทิศทางนโยบายใน 4 มุม
plt.text(35000, 0.85, "กลุ่ม 1: รวยแต่เหลื่อมล้ำสูง\n(ควรมุ่งลดช่องว่างรายได้)", color='darkred', fontsize=16, alpha=0.7)
plt.text(35000, 0.35, "กลุ่ม 2: รวยและเท่าเทียม\n(รักษามาตรฐาน)", color='darkgreen', fontsize=16, alpha=0.7)
plt.text(15000, 0.85, "กลุ่ม 3: จนและเหลื่อมล้ำสูง\n(วิกฤต: ต้องแก้ทั้งสองด้าน)", color='purple', fontsize=16, alpha=0.7)
plt.text(15000, 0.35, "กลุ่ม 4: จนแต่เท่าเทียม\n(ควรเน้นยกระดับรายได้โดยรวม)", color='darkorange', fontsize=16, alpha=0.7)


plt.legend()
plt.tight_layout()
plt.show()
```

**จากกราฟscatter plot ที่เราplot รายได้เฉลี่ยและ CV** <br>**ทั้ง 4 กลุ่ม**<br>กลุ่ม1. จันทบุรี เป็นกลุ่มที่มีทั้งรายได้สูงและ cv สูงเนื่องจากเป็นจังหวัดที่ทำเกี่ยวกับการเกษตรทำให้รายได้ไปกระจุกตัวอยู่ที่การเกษตรทำให้บางครัวเรือนมีรายได้สูงบางครัวเรือยมีรายได้ตำ่กว่าทำให้มีการเหลื่อมลำ้ได้<br>กลุ่มที่2.ปทุมธานี เป็นกลุ่มที่รายได้เฉลี่ยสูงแต่มี CV ตำ่ถือว่าเป็นกลุ่มที่สมดุลเนื่องจากเป็นจังหวัดปริมณฑลและมีทั้งภาคอุตสาหกรรมและ การบริการทำให้รายได้ค่อนหลากหลายทำให้เกิดการกระจายตัวทางรายได้ที่สมดุลกัน<br>กลุ่มที่3.นราธิวาสเป็นกลุ่มที่มีรายได้เฉลี่ยกับ CV ตำ่ สะท้อนถึงปัญหาทางด้านเศรษฐกิจและการประกอบอาชีพ<br>กลุ่มที่4.เชียงราย เป็นกลุ่มที่รายได้เฉลี่ยตำ่แต่ CV สูง เนื่องจากการกระจายตัวของรายได้ไม่เท่ากันเกิดจากความแตกต่างของแหล่งรายได้
